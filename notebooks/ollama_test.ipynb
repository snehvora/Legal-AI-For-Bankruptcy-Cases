```python
import ollama
```

```python
response = ollama.chat(
    model='gemma2:2b',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
)
```

```python
response['message']['content']
```

```python
stream = ollama.chat(
    model='gemma2:2b',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)
```

```python
for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```
